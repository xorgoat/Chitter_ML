### Split data into sets

```python
# Move files over to correct folders

import os
from os.path import isdir
import math
import shutil

path = "/content/SE/_final_insect_images/"
print(os.listdir(path))
lst = os.listdir(path + "species/")

for species in lst: 
  if not isdir(path + "validation/" + species):
    os.mkdir(path + "validation/" + species)
  if not isdir(path + "test/" + species):  
    os.mkdir(path + "test/" + species)
  if not isdir(path + "train/" + species):
    os.mkdir(path + "train/" + species)

  counter = 0
  sp_lst = os.listdir(path + "species/" + species)
  num_rec = len(sp_lst)
  
  #divide up files into 80 10 10 distribution (round up in favor of test and val so there is at least one file for each folder)
  num_val = num_test = math.ceil(0.1 * num_rec)
  num_train = num_rec - 2 * num_test
  #move files into correct folders
  while counter < num_rec:
    if counter < num_val:
      try:
        shutil.copy(path + "species/" + species + "/" + sp_lst[counter], path + "validation/" + species + "/" + sp_lst[counter])
      except:
        print("file already moved")
    elif counter < (num_val + num_test):
      try:
        shutil.copy(path + "species/" + species + "/" + sp_lst[counter], path + "test/" + species + "/" + sp_lst[counter])
      except:
        print("file already moved")
    else:
      try:
        shutil.copy(path + "species/" + species + "/" + sp_lst[counter], path + "train/" + species + "/" + sp_lst[counter])
      except:
        print("file already moved")
    counter += 1
    
```

# **Model**

Creating a simple model for Bug Images

```python
#Model creation and selection
param = {'n_neighbors':[1, 10, 100, 500]}

grid_search_knn = GridSearchCV(KNeighborsClassifier(), param_grid = param, cv=5)
grid_search_knn.fit(X_train_pca, y_train)

#Find the best hyperparameter
print("Best parameters: {}".format(grid_search_knn.best_params_))
```

Running the simple model for Bug Images

```python
knn = KNeighborsClassifier(n_neighbors = 500).fit(X_train_pca, y_train)
y_pred_knn = knn.predict(X_test_pca)
```

Evaluating the simple model for Bug Images

```python
#evaluation metrics
print("Accuracy: {:.3f}".format(accuracy_score(y_test, y_pred_knn)))
print("Confusion matrix:\n{}".format(confusion_matrix(y_test, y_pred_knn)))
print("Weighted f1-score:", f1_score(y_test, y_pred_knn, average='weighted'))
```
